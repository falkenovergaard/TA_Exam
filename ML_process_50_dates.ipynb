```python
import pandas as pd
```

```python
'ML_Data/DJI_vader.csv'

def merge_50_first_with_price_date(df_type):
    
    
    if df_type == "DJI_vader":
        df = pd.read_csv('ML_Data/DJI_vader.csv')
        
        
    df=df.drop(['Unnamed: 0', 'label'], axis=1)
    df=df.iloc[:50]
    
    
    if df_type == "DJI_vader":
        df_price = pd.read_excel('data/price/DJI.xlsx',skiprows=[0])
    
    df_price['date'] = pd.to_datetime(df_price['Date'])
    df['date'] = pd.to_datetime(df['date'])
    

    df = pd.merge(df, df_price, how='inner', on='date')
    
    
    df = df.drop(['High', 'Low','Upper bound', 'Lower Bound','Date'], axis=1)
    
    #df = pd.merge(df,df_price, left_on="date", right_on="Date",how="inner")
    
    
    #BTC_neg
    #BTC_pos
    #BTC_vader
    
    #DJI_neg
    #DJI_pos
    
    #TSLA_neg
    #TSLA_pos
    #TSLA_vader
    
    return df, df_type
    
    

result = merge_50_first_with_price_date("DJI_vader")
result[0]
```
