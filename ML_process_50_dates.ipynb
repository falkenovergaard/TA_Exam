```python
import pandas as pd
```

```python

def merge_50_first_with_price_date(df_type, df_label):
    
    
    df = pd.read_csv('ML_Data/'+ df_type +'.csv')
        
        
    df=df.drop(['Unnamed: 0', 'label'], axis=1)
    df=df.iloc[:50]
    
    
    if df_type[0:3]=="DJI":
        df_price = pd.read_excel('data/price/'+df_type[0:3]+'.xlsx',skiprows=[0])
        
    elif df_type[0:3]=="BTC":
        df_price = pd.read_excel('data/price/'+df_type[0:3]+'.xlsx',skiprows=[0])
        
    else:
        df_price = pd.read_excel('data/price/TSLA.xlsx',skiprows=[0])
        
    
    df_price['date'] = pd.to_datetime(df_price['Date'])
    df['date'] = pd.to_datetime(df['date'])
    df_label['date'] = pd.to_datetime(df_label['date'])
    

    df = pd.merge(df, df_price, how='inner', on='date')
    
    df = pd.merge(df, df_label, how='inner', on='date')
 
    df = df[['date','Close','Label']]
   
    #df = pd.merge(df,df_price, left_on="date", right_on="Date",how="inner")
    
    
    #BTC_neg
    #BTC_pos
    #BTC_vader
    
    #DJI_neg
    #DJI_pos
    
    #TSLA_neg
    #TSLA_pos
    #TSLA_vader
    
    return df, df_type
    

```

```python
DJI_vader_labled = pd.read_csv("ML_Data/DJI_vader_50_labeled.csv")
```

```python

result_DJI_vader = merge_50_first_with_price_date("DJI_vader",DJI_vader_labled)
result_DJI_vader[0].to_csv("ML_Data/first_50_dates/"+result_DJI_vader[1]+"_50.csv") 
result_DJI_vader[0]
```
